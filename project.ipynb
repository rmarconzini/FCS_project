# Exercise 13
Build a new dataframe (called common) where, for each pickup location we keep all trips to the 5 most common destinations (i.e. each pickup location can have different common destinations).

```python
trip_counts = (
    df.groupby(["PULocationID", "DOLocationID"])
    .size()
    .reset_index(name="trip_count")
)

trip_counts["rank"] = trip_counts.groupby("PULocationID")["trip_count"].rank(
    method="first", ascending=False
)

top_destinations = trip_counts[trip_counts["rank"] <= 5]

common = df.merge(
    top_destinations[["PULocationID", "DOLocationID"]],
    on=["PULocationID", "DOLocationID"],
    how="inner",
)

display(common)
```

# Exercise 14

# Exercises 15
Compute the difference of the average fare amount computed in the previous point with those computed at point 9.

```python
df_avg_fare = df.groupby(["payment_type", "pickup_interval"]).agg(
    avg_fare_amount=("fare_amount", "mean")
).reset_index()

common_avg_fare = common.groupby(["payment_type", "pickup_interval"]).agg(
    avg_fare_amount=("fare_amount", "mean")
).reset_index()

merged_avg_fare = pd.merge(
    df_avg_fare, common_avg_fare, on=["payment_type", "pickup_interval"], suffixes=("_df", "_common")
)

merged_avg_fare["fare_difference"] = merged_avg_fare["avg_fare_amount_common"] - merged_avg_fare["avg_fare_amount_df"]

display(merged_avg_fare)
```

# Exercise 16
Compute the ratio between the differences computed in the previous point and those computed in point 9. Note: you have to compute a ratio for each pair (payment type, interval).


```python
df_diff = df_avg_fare.groupby(["payment_type", "pickup_interval"]).agg(
    fare_difference_df=("avg_fare_amount", "mean")
).reset_index()

merged_diff = pd.merge(
    merged_avg_fare, df_diff, on=["payment_type", "pickup_interval"]
)

merged_diff["ratio"] = merged_diff["fare_difference"] / merged_diff["fare_difference_df"]

# Output the result
mask = ["payment_type", "pickup_interval", "ratio"]
display(merged_diff[mask])
```

# Exercise 17
Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.


Steps to Build Chains:
-   Sort the data by VendorID, PULocationID, DOLocationID and tpep_dropoff_datetime.
-   Check consecutive trips: Compare each trip with the next one to see if they satisfy the chaining conditions.
-   Assign chain IDs: Use a new column chain to group trips into chains.

```python
df = df.sort_values(by=["VendorID", "tpep_dropoff_datetime"]).reset_index(drop=True)


df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])
df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])


# Initialize a chain column with default value -1
df["chain"] = -1

current_chain = 0
for i in tqdm.tqdm(range(len(df))):
    if df.loc[i, "chain"] != -1:
        continue

    current_chain += 1
    df.loc[i, "chain"] = current_chain

    for j in tqdm.tqdm(range(i + 1, len(df))):
        if (
            df.loc[j, "VendorID"] == df.loc[i, "VendorID"]
            and df.loc[j, "PULocationID"] == df.loc[i, "DOLocationID"]
            and df.loc[j, "tpep_pickup_datetime"] > df.loc[i, "tpep_dropoff_datetime"]
            and (df.loc[j, "tpep_pickup_datetime"] - df.loc[i, "tpep_dropoff_datetime"]).total_seconds() <= 120
        ):
            df.loc[j, "chain"] = current_chain
            i = j

display(df)
```
