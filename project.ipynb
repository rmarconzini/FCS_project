```python
import pandas as pd
```

# Exercise 1
Extract all trips with trip_distance larger than 50

```python
long_trips = df.loc[df.trip_distance > 50]

print("Total number of trip with distance > 50 miles: ", long_trips.shape[0])

display(long_trips.head())
```

# Exercise 2
Extract all trips where payment_type is missing

```python
missing_payment = df.loc[df.payment_type.isna()]

print("Total number of trips with missing payment type: ", missing_payment.shape[0])

display(missing_payment.head())
```

# Excercise 3

For each (PULocationID, DOLocationID) pair, determine the number of trips

```python
number_of_trips = df.groupby(
    ["PULocationID", "DOLocationID"]
).size().reset_index(name="Number of Trips")

number_of_trips.head()
```

# Exercise 4

Save all rows with missing VendorID, passenger_count, store_and_fwd_flag, payment_type in a new dataframe called bad, and remove those rows from the original dataframe.

```python
mask = (df.VendorID.isna()) & (df.passenger_count.isna()) & (df.store_and_fwd_flag.isna()) & (df.payment_type.isna())

bad = df.loc[mask]

print("Total number of rows: ", bad.shape[0])

display(bad.head())
```

# Exercises 5
Add a duration column storing how long each trip has taken (use tpep_pickup_datetime, tpep_dropoff_datetime)

```python
df["duration"] = pd.to_datetime(df.tpep_dropoff_datetime) - pd.to_datetime(df.tpep_pickup_datetime)
```
